```python
import pandas as pd
```

```python
df = pd.read_csv('../data/gapminder.tsv', sep='\t')
```

```python
df.head()
```

```python
df.groupby('year')['lifeExp'].mean()
```

```python
df.groupby('year')['lifeExp'].describe()
```

```python
import numpy as np
df.groupby('year')['lifeExp'].agg(np.mean)
```

```python
def my_mean(values):
    n = len(values)
    sum = 0
    for value in values:
        sum += value
    return sum / n
```

```python
df.groupby('year')['lifeExp'].agg(my_mean)
```

```python
(df
 .groupby('year')['lifeExp']
 .agg([np.count_nonzero, np.mean, np.std, my_mean])
 .reset_index()
)
```
